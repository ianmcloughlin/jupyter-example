```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (20, 12)
```

```python
import networkx as nx
```

```python
import math
```

```python
import numpy as np
```

```python
rng = np.random.default_rng(seed=0)
```

```python
L = list(rng.permutation(10))
L
```

```python
T = nx.balanced_tree(2, math.floor(math.log(len(L), 2)))
T.remove_nodes_from(list(T.nodes)[len(L):])
T = nx.relabel_nodes(T, dict(zip(T.nodes, L)), copy=True)

pos = nx.drawing.nx_pydot.graphviz_layout(T, prog="dot")
nx.draw(T, pos, with_labels=True, node_color='lightgrey', node_size=800, font_size=18)
nx.draw_networkx(T.subgraph(list(T.nodes)[4]), pos=pos, node_color='pink', node_size=800, font_size=18)
```
